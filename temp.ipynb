```python
def calc_stock_return(buy_price, current_price, shares=1):
    cost = buy_price * shares
    value = current_price * shares
    profit = value - cost
    rate = profit / cost * 100

    return {
        "cost": cost,
        "value": value,
        "profit": profit,
        "rate": rate
    }
 
# 示例
ret = calc_stock_return(2.2106, 2.1833, 49.76)
print(f"成本: {ret['cost']:.2f}")
print(f"当前市值: {ret['value']:.2f}")
print(f"盈亏额: {ret['profit']:.2f}")
print(f"盈亏率: {ret['rate']:.2f}%")
```
