```python
def calc_stock_return(buy_price, current_price, shares=1):
    cost = buy_price * shares
    value = current_price * shares
    profit = value - cost
    rate = profit / cost * 100

    return {
        "cost": cost,
        "value": value,
        "profit": profit,
        "rate": rate
    }
 
# 示例
ret = calc_stock_return(2.2106, 2.1833, 49.76)
print(f"成本: {ret['cost']:.2f}")
print(f"当前市值: {ret['value']:.2f}")
print(f"盈亏额: {ret['profit']:.2f}")
print(f"盈亏率: {ret['rate']:.2f}%")
```

```python
from tqdm import tqdm
import json
import numpy as np

file_list_path = '/mnt/data/zhengwenhao/workspace/cad_graph/ProxBFN-main/data/deepcad/train_all_graphs.txt'
metrics_path = '/mnt/data/zhengwenhao/workspace/cad_graph/ProxBFN-main/data/deepcad/cad_metrics.json'

# Load the entire metrics data (train, validation, and test)
with open(metrics_path, "r") as fp:
    all_metrics_data = json.load(fp)

# Combine all the sections (train, validation, and test)
train_data = all_metrics_data['train']
validation_data = all_metrics_data.get('validation', {})
test_data = all_metrics_data.get('test', {})

with open(file_list_path, "r") as f:
    graph_rel_paths = [line.strip() for line in f if line.strip()]

# Step 1: Collect all area and vol values
area_values = []
vol_values = []

for data_id in tqdm(graph_rel_paths, desc="Collecting area and vol values"):
    metrics_data = train_data.get(data_id) or validation_data.get(data_id) or test_data.get(data_id)
    area_values.append(metrics_data['area'])
    vol_values.append(metrics_data['vol'])

# Step 2: Calculate mean and standard deviation for area and vol
# Convert to numpy arrays for easier calculations
area_values = np.array(area_values)
vol_values = np.array(vol_values)

# Calculate mean and standard deviation for area and vol
mean_area = np.mean(area_values)
std_area = np.std(area_values)
mean_vol = np.mean(vol_values)
std_vol = np.std(vol_values)

print(f"Area - Mean: {mean_area}, Std Dev: {std_area}")
print(f"Volume - Mean: {mean_vol}, Std Dev: {std_vol}")
```
